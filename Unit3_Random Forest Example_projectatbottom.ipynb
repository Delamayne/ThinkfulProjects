```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

```

```python
y2015['id'] = pd.to_numeric(y2015['id'], errors='coerce')
y2015['int_rate'] = pd.to_numeric(y2015.int_rate.str.strip('%'), errors='coerce')

y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state', 'desc'], 1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
np.random.seed(123)
```

```python
label_encoder = LabelEncoder()
```

```python
y_encoded = label_encoder.fit_transform(y2015['loan_status']).astype('float64')
```

```python
y_e = pd.DataFrame()
y_e['outcome_encoded'] = y_encoded
y_e.head()
```

```python
X.head()
```

```python
model = sm.OLS(y_e, X).fit()
```

```python
pDF = pd.DataFrame(data=model.pvalues)
```

```python
pDF.reset_index(inplace = True)
pd.options.display.float_format = '{:.4f}'.format
pDF.rename({"index":"feature",0:"p_val"}, axis="columns", inplace=True)
```

```python
pDF.columns
```

```python
pDF = pDF[pDF.p_val < .000000000001]
final_f = pDF.feature.tolist()
X_final = X[final_f]
#X_final = X_final.drop(columns=['id'])
pDF.describe()
```

```python
rfc_2 = ensemble.RandomForestClassifier()

cross_val_score(rfc_2, X_final, Y, cv=10)
```
