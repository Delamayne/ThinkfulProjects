```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
batting_avg = np.random.beta(81, 219, 100)
```

```python
plt.hist(batting_avg)
plt.axvline(batting_avg.mean(), color = 'orange', linestyle = 'solid', linewidth = 2)
plt.axvline(batting_avg.mean() + batting_avg.std(), color = 'orange', linestyle = 'dashed', linewidth = 2)
plt.axvline(batting_avg.mean() - batting_avg.std(), color = 'orange', linestyle = 'dashed', linewidth = 2)
plt.show()
#Batting Average Distribution 1/6
```

```python
batting_avg.describe()
```

```python
gumbel = np.random.gumbel(0,.5,100)
plt.hist(gumbel)
plt.axvline(gumbel.mean(), color = 'black', linestyle = 'solid', linewidth = 3)
plt.axvline(gumbel.mean()+gumbel.std(), color = 'black', linestyle = 'dashed', linewidth = 3)
plt.axvline(gumbel.mean()-gumbel.std(), color = 'black', linestyle = 'dashed', linewidth = 3)
plt.show()
# Gumble distribution, right-skewed, mean > peak, 2/6
```

```python
triangular = np.random.triangular(-2,2,10,100)
plt.hist(triangular)
plt.axvline(triangular.mean(), color = 'green', linewidth = 2, linestyle = 'solid')
plt.axvline(triangular.mean()+triangular.std(), color = 'green', linewidth = 2, linestyle = 'dashed')
plt.axvline(triangular.mean()-triangular.std(), color = 'green', linewidth = 2, linestyle = 'dashed')
plt.show()
# not sure what this would be used for, can potentially be normal or skewed dist
```

```python
uniform = np.random.uniform(0,25,100)
plt.hist(uniform)
plt.axvline(uniform.mean(), color = 'orange', linestyle = 'solid', linewidth = 2)
plt.axvline(uniform.mean()+uniform.std(), color = 'orange', linestyle = 'dashed', linewidth = 2)
plt.axvline(uniform.mean()-uniform.std(), color = 'orange', linestyle = 'dashed', linewidth = 2)
plt.show()
#testing to see what this one does, have a few Q's

```

```python
gamma = np.random.gamma(2, 1, 100)
plt.hist(gamma)
plt.axvline(gamma.mean(), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(gamma.mean() + gamma.std(), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(gamma.mean() - gamma.std(), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.show()
#testing this vs standard_gamma
```

```python
standard_gamma = np.random.standard_gamma(2,100)
plt.hist(standard_gamma)
plt.axvline(standard_gamma.mean(), color = 'yellow', linestyle = 'solid', linewidth = 3)
plt.axvline(standard_gamma.mean() + standard_gamma.std(), color = 'yellow', linestyle = 'dashed', linewidth = 3)
plt.axvline(standard_gamma.mean() - standard_gamma.std(), color = 'yellow', linestyle = 'dashed', linewidth = 3)
plt.show()
#synonymous with gamma where scale parameter is the same?

```

```python
# Additional Project
var_combine = pd.DataFrame()
var_combine['a'] = np.random.normal(5,.5,100)
var_combine['b'] = np.random.normal(10,1,100)
var_combine['c'] = var_combine.a + var_combine.b
```
