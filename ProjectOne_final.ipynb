```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('UNWindEnergy.csv')
```

```python
df
```

```python
#look at last year with data across many countries
df['Year'].value_counts()
```

```python
#add a '1' if year is between 2011-2016
df['valid_year'] = np.where(np.logical_and(df.Year >= 2011, df.Year <= 2016), 1, 0)
```

```python
#add dataframe with unique country list and country count, where country count = 6

df2 = pd.DataFrame()
df2['country_count'] = df.groupby('Country or Area')['valid_year'].sum()
df2=df2[df2.country_count == 6]
df2
```

```python
#merge original df with df2, only getting countries with valid years of data
df2 = pd.merge(df2, df, left_index = True, right_on = 'Country or Area')
df2
```

```python
#remove years outside of 2011-2016
df2=df2[df2.valid_year == 1]
df2
```

```python
df2 = df2[['Country or Area', 'Year', 'Quantity']]
df2
```

```python
df3=df2[df2['Year']==2016] 
df4=df2[df2['Year']==2011]
df3
```

```python
#get rid of year, how to simplify this when creating df in previous step?
df4=df4[['Country or Area', 'Quantity']]
df4.columns = ['Country or Area', 'Old_Quantity']
df4
```

```python
#new dataframe just so I can play around w merge parameters without changing original
dffinal = pd.merge(df3, df4, left_on = 'Country or Area', right_on = 'Country or Area')
dffinal
```

```python
#do calculations, remove countries where either current or old quantity = 0; removes two countries
dffinal['Percent_Change'] = np.round(((dffinal.Quantity / dffinal.Old_Quantity) -1)*100, decimals = 0)
dffinal['Nominal_Change'] = np.round((dffinal.Quantity - dffinal.Old_Quantity), decimals = 0)
dffinal = dffinal[dffinal['Quantity'] > 0]
dffinal = dffinal[dffinal['Old_Quantity'] > 0]
dffinal
```

```python
#do some 'top lists', visuals, and summary stats
dffinal.nlargest(10, 'Percent_Change')
```

```python
dffinal.nlargest(10, 'Nominal_Change')
```

```python
median_change = np.median(dffinal.Percent_Change)
print('The Typical Country Experienced a {}% five-year cumulative change in Wind Production.'.format(median_change))
```
