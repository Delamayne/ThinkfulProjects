```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
apt_data = pd.read_csv('denver_apt_class_rg_vacancy.csv')
```

```python
#choose two continous variables, and plot them three different ways
sns.set(style="white")
#fig, ax = plt.subplots()
#fig.set_size_inches(4,12)

#***how do I show cross at 0%?

g = sns.lmplot(y='rent_growth',
               x='occupancy_rate',
               data=apt_data,
               fit_reg=False,
               #why brackets here?
               scatter_kws={'alpha':0.6})
g.set_ylabels("Rent Growth Y/Y 2018")
g.set_xlabels("Average Occupancy Rate in 2018")
plt.title('Rent Growth vs. Occupancy: Large Denver Apartments')
#
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
#plt.gca().set_xticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 

plt.show()



g = sns.lmplot(y='rent_growth',
               x='occupancy_rate',
               data=apt_data,
               fit_reg=True,
               height=8,
               aspect=8/16,
               #why brackets here?
               scatter_kws={'alpha':0.6})
g.set_ylabels("Rent Growth Y/Y 2018")
g.set_xlabels("Average Occupancy Rate in 2018")

plt.title('Rent Growth vs. Occupancy: Large Denver Apartments')
#
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
#plt.gca().set_xticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
#plt.figure(figsize=(4,8))
plt.show()

g = sns.lmplot(y='rent_growth',
               x='occupancy_rate',
               data=apt_data,
               fit_reg=True,
               lowess=True,
               #why brackets here?
               scatter_kws={'alpha':0.6})
g.set_ylabels("Rent Growth Y/Y 2018")
g.set_xlabels("Average Occupancy Rate in 2018")
plt.title('Rent Growth vs. Occupancy: Large Denver Apartments')
#
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
#plt.gca().set_xticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 

plt.show()


#not a good way to visualize, will do in next exercise
g = sns.lmplot(y='rent_growth',
               x='occupancy_rate',
               hue='class',
               data=apt_data,
               fit_reg=True,
               lowess=False,
               ci=False,
                 height=8,
               aspect=8/16,
               #why brackets here?
               scatter_kws={'alpha':0.6})
g.set_ylabels("Rent Growth Y/Y 2018")
g.set_xlabels("Average Occupancy Rate in 2018")
plt.ylim(-.15,.2)
plt.title('Rent Growth vs. Occupancy: Large Denver Apartments')
#
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
#plt.gca().set_xticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 

plt.show()
```
